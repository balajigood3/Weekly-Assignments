```python
#1.
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

X = [[1, 2], [3, 4], [5, 6]]
y = [1, 0, 1]

rfe = RFE(estimator=LogisticRegression(), n_features_to_select=1)
X_new = rfe.fit_transform(X, y)
print(X_new)
```

```python
#2.
from sklearn.feature_selection import VarianceThreshold

X = [[0, 1, 4], [1, 2, 1], [4, 2, 5]]
selector = VarianceThreshold(threshold=0.1)
X_new = selector.fit_transform(X)
print(X_new)
```

```python
#3.
from sklearn.metrics.pairwise import cosine_similarity
data = [[5, 0, 3], [0, 0, 0]]
similarity = cosine_similarity(data)
print(similarity)
```

```python
#4.
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeClassifier

X = [[1, 2], [3, 4], [5, 6]]
y = [1, 0, 1]
selector = SelectFromModel(DecisionTreeClassifier(), threshold="mean")
X_new = selector.fit_transform(X, y)
print(X_new)
```

```python
#5.
from pandas import DataFrame, pivot_table, pivot
from pandas import pandas as pd
data = pd.DataFrame({
    'user': ['A', 'B', 'A'],
    'item': ['I1', 'I2', 'I2'],
    'rate': [5, 4, 3]
})
pivot = data.pivot_table(
    index='user',
    columns='item',
    values='rate',
    aggfunc='mean'  # or 'sum', 'max', etc.
)
print(pivot)
```

```python
#6.
from sklearn.decomposition import PCA
import numpy as np

X = np.array([[1, 2], [3, 4]])
pca = PCA(n_components=1)
X_new = pca.fit_transform(X)
print(X_new)
```

```python
#7.
ratings = pd.DataFrame({
    'User1': [5, None, 3],
    'User2': [None, 4, None],
    'User3': [1, 2, None]
})
ratings = ratings.fillna(0)
sim = cosine_similarity(ratings.T)
print(sim)
```

```python
#8.
import numpy as np

user_vec = np.array([[1, 2]])
item_vec = np.array([[0.5], [1.0]])
score = np.dot(user_vec, item_vec)
print("Predicted Score:", score)
```

```python
#9.
from sklearn.neighbors import NearestNeighbors
import numpy as np

X = np.array([[1, 0, 1], [0, 1, 0], [1, 1, 1]])
model = NearestNeighbors()
model.fit(X)
distances, indices = model.kneighbors([X[0]], n_neighbors=2)
print("Neighbours indicates:", indices)
```

```python
#10.
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

a = np.array([1, 0, 1]).reshape(-1, 1)
b = np.array([0, 1, 1]).reshape(-1, 1)

sim = cosine_similarity(a, b)
print("Similarity:", sim)
```
