```python
#4.Write a Python function to find the factorial of 5. 
#Expected Output: 120
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Calling the function to find the factorial of 5
factorial_of_5 = factorial(5)
print(factorial_of_5)
```

```python
#5.Write a Python program to check if a number is prime. 
#Expected Output (for input 7): 7 is a prime number
num = 7

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```
