```python
# Value Error
from gensim import corpora, models
texts = [["cat", "dog", "mouse"], ["dog", "lion"]]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
lda = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=2)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Example text
text = "Great life due to AI"

# Generate the word cloud
wc = WordCloud(width=800, height=400, background_color='white').generate(text)

# Convert to PIL Image first, then display
plt.figure(figsize=(8, 4))
plt.imshow(wc.to_image(), interpolation='bilinear') 
plt.axis('off')
plt.title("Word Cloud Output", fontsize=16)
plt.show()
```
