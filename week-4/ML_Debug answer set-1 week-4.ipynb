```python
#8.Replaced NaN with column mean using SimpleImputer to allow LinearRegression to train and predict correctly.import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Creating a dataset with missing values
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

# Handling missing values by replacing NaN with the column mean
imputer = SimpleImputer(strategy="mean")
X_train_imputed = imputer.fit_transform(X_train)

# Train the model
model = LinearRegression()
model.fit(X_train_imputed, y_train)

# Make a prediction
prediction = model.predict([[3, 4]])
print("Predicted Output:", prediction)  
```

```python
#10.Explanation:SVM is sensitive to feature scales. StandardScaler normalizes the features, improving model performance.from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train = [[1, 100], [2, 200], [3, 300]]
y_train = [0, 1, 0]
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
model = SVC()
model.fit(X_train_scaled, y_train)
X_test = [[1.5, 150], [2.5, 250]]
X_test_scaled = scaler.transform(X_test)
predictions = model.predict(X_test_scaled)
print(predictions)
```
