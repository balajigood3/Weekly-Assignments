```python
#1. Prompt user until they enter a valid integer
#Question: A user is required to enter a valid number in a form, but users sometimes input invalid data. Write logic to repeatedly prompt the user until they enter a valid integer.
while True:
    user_input = input("Enter a valid integer: ")
    try:
        number = int(user_input)
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")

print("You entered:", number)
```

```python
#2. Find the most frequently occurring number in a list
#Question: A data analysis tool processes a list of numbers and needs to identify the most frequently occurring value. Write logic to find the most frequently occurring number in a given list.

from collections import Counter

numbers = [1, 2, 2, 3, 3, 3, 4]
frequency = Counter(numbers)
most_common = frequency.most_common(1)[0][0]

print("Most frequent number:", most_common)
```

```python
#3.A text-processing application needs to compare words and check if they are anagrams (contain the same letters in a different order). 
#Write logic to determine whether two given strings are anagrams.

def are_anagrams(str1, str2):
    return sorted(str1.replace(" ", "").lower()) == sorted(str2.replace(" ", "").lower())

print(are_anagrams("listen", "silent"))  # True
```

```python
#4.A speech analysis program needs to count the number of vowel sounds in a given input.
#Write logic to count the number of vowels in a given string.

text = "Hello World"
vowels = "aeiouAEIOU"
count = sum(1 for char in text if char in vowels)

print("Number of vowels:", count)
```

```python
#5.A text-editing software includes a feature to reverse the order of words in a sentence for stylistic effects. 
#Write logic to reverse the order of words in a sentence while keeping the words themselves intact.

sentence = "Python is fun"
reversed_sentence = " ".join(sentence.split()[::-1])

print(reversed_sentence)  # Output: "fun is Python"
```

```python
#6.A missing number is detected in a sequence of values stored in a database.
#Write logic to find the missing number in a list containing n-1 numbers from 1 to n.

numbers = [1, 2, 4, 5]  # Missing 3
n = len(numbers) + 1
expected_sum = n * (n + 1) // 2
missing_number = expected_sum - sum(numbers)

print("Missing number:", missing_number)
```

```python
#7.An ATM machine processes withdrawal requests and needs to ensure that users cannot withdraw more than their account balance. 
#Write logic to allow a withdrawal only if the balance is sufficient.

balance = 1000
withdrawal = int(input("Enter withdrawal amount: "))

if withdrawal <= balance:
    balance -= withdrawal
    print("Withdrawal successful. Remaining balance:", balance)
else:
    print("Insufficient balance!")
```

```python
#8.A system needs to verify whether a given dataset contains duplicate entries.
#Write logic to check whether a given list contains duplicate values.

numbers = [1, 2, 3, 2, 4]
has_duplicates = len(numbers) != len(set(numbers))

print("Contains duplicates:", has_duplicates)
```

```python
#9.A digital calculator includes a feature to sum the digits of a number for verification purposes. 
# Write logic to calculate the sum of all digits in a given integer.

num = 12345
digit_sum = sum(int(d) for d in str(num))

print("Sum of digits:", digit_sum)
```

```python
#10.A language-learning app wants to verify whether a given sentence is a pangram (contains every letter of the alphabet at least once). 
# Write logic to check if a given sentence is a pangram.

import string

sentence = "The quick brown fox jumps over the lazy dog"
alphabet = set(string.ascii_lowercase)
is_pangram = alphabet <= set(sentence.lower())

print("Is pangram:", is_pangram)
```
