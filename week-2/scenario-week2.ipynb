```python
#5.A text-processing tool reverses a given word or sentence for formatting purposes.
#Write logic to take a word or sentence as input and produce its reversed version.

text = "RAMKUMAR.P"

def reverse_print(text):
    reversed_text = text[::-1]
    print(reversed_text)

reverse_print(text)
```

```python
# A calendar system verifies whether a given year is a leap year based on standard leap year rules.
# Write logic to determine whether a given year is a leap year.
Leap_Year = "4 Years once is Leap Year"
Normal_year = "Not a Leap Year"

year = int(input("Enter the year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} - {Leap_Year}")
else:
    print(f"{year} - {Normal_year}")
```
