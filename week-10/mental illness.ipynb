```python
print("=== Mental Wellness Daily Check-In ===")

mood_log = []

while True:
    print("\nOptions:")
    print("1. Enter today's mood (1-5)")
    print("2. View weekly summary")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        mood = int(input("Rate your mood today (1-5): "))
        if 1 <= mood <= 5:
            mood_log.append(mood)
            print("Mood recorded! ðŸŒŸ")
        else:
            print("Please enter a number between 1 and 5.")

    elif choice == "2":
        if mood_log:
            average = sum(mood_log) / len(mood_log)
            print("Your weekly mood log:", mood_log)
            print(f"Average mood: {average:.2f}")
        else:
            print("No moods recorded yet.")

    elif choice == "3":
        print("Exiting Wellness Check-In. Take care! ðŸ’›")
        break

    else:
        print("Invalid choice, try again.")
```
